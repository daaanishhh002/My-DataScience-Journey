# Matplotlib

Matplotlib is a comprehensive data visualization library for Python. It provides a flexible and high-quality platform for creating static, animated, and interactive plots in Python. Matplotlib is widely used for producing a variety of plots and charts, making it a fundamental tool for data scientists, researchers, and analysts.

---

Key features of Matplotlib include:
1. **Simple and Intuitive Interface:**
   - Matplotlib provides a simple and consistent interface for creating a wide range of plots, allowing users to customize the appearance of their visualizations.
2. **Versatility:**
   - It supports a variety of plot types, including line plots, scatter plots, bar plots, histograms, pie charts, 3D plots, and more. This versatility makes it suitable for a broad range of data visualization tasks.
3. **Customization:**
   - Matplotlib allows users to customize almost every aspect of a plot, including colors, line styles, markers, fonts, and labels. This flexibility enables the creation of highly polished and publication-quality visualizations.
4. **Integration with Jupyter Notebooks:**
   - Matplotlib seamlessly integrates with Jupyter Notebooks, providing an interactive environment for visualizing data and immediately seeing the results.
5. **Extensibility:**
   - Matplotlib is highly extensible, allowing users to build upon its functionality by incorporating additional libraries and toolkits, such as Seaborn and Basemap.
6. **Backends:**
   - It supports different backends for rendering plots, allowing users to generate plots in various formats, including interactive GUIs, static images, or embedded in web applications.

Matplotlib is often used in conjunction with NumPy, Pandas, and other data science libraries to visualize and analyze data. The library is well-documented, and the official Matplotlib website provides extensive galleries, tutorials, and examples to help users get started with creating effective visualizations.

## Line Plot

```python
x = np.arange(50,550,15)
y = np.sin(x)

plt.plot(x,y)
plt.title('Some Data')
plt.xlabel('Input')
plt.ylabel('Output');
```

```python
x = np.arange(50,550,15)
y = np.sin(x)
z = np.tan(x)

plt.plot(x,y,
         label='$sin(x)$',
         lw=3)
plt.plot(x,z,
         'D--',
         label='$tan(x)$')
plt.legend();
```

```python
x = np.arange(50,550,20)
y = np.sin(x+5)
z = np.cos(x-5)

plt.plot(x,y,'o--',label='sin(x)',markersize=4)
plt.plot(x,z,'*:r',label='cos(x)')
plt.legend(loc=1,bbox_to_anchor=(1.25,1));
```

## Scatter Plot

```python
x = rng.normal(300,10,50)
y = rng.normal(200,10,50) * 0.6
z = rng.normal(100,5,50)

plt.scatter(x,y,marker='*',color='purple')
plt.scatter(x,z,marker='^',color='green')
plt.title('Some Graph',fontdict={'fontsize':25, 'color':'maroon'})
plt.xlabel('X-axis',fontsize=13)
plt.ylabel('Y-axis',fontsize=13);
```

```python
x = rng.normal(300,20,500)
y = rng.normal(200,50,500)
z = rng.normal(100,50,500)

plt.scatter(x,y,marker='*',color='purple',alpha=0.4,label='idk')
plt.scatter(y,z,marker='^',color='green',alpha=0.4,label='idk again.')
plt.title('Some Graph',fontdict={'fontsize':25, 'color':'maroon'})
plt.xlabel('X-axis',fontsize=13)
plt.ylabel('Y-axis',fontsize=13)
plt.legend(bbox_to_anchor=(0.3,0.75));
```

## Bar Plot

```python
x = ['SUVs','Sedans','Hatchbacks','Coupes','Supercars','Hypercars']
y = np.array([50_000,41_000,35_000,20_000,10_000,5_000])
color=['purple','red','blue','yellow','orange','maroon']

plt.bar(x,y,width=.7,color=color)
```

```python
x = ['SUVs','Sedans','Hatchbacks','Coupes','Supercars','Hypercars']
y = np.array([50_000,41_000,35_000,20_000,10_000,5_000])
z = np.array([35_000,24_100,13_500,12_000,8_000,500])

plt.bar(x,y,label='Pre 2020')
plt.bar(x,z,label='In 2020')
plt.legend(bbox_to_anchor=(0.8,0.75),title='Era');
```

## Histograms

```python
x = rng.normal(100,15,10_000)

plt.hist(x,bins=20)
plt.title('Some Graph',fontsize='20',fontweight='bold')
plt.xlabel('IQ Score')
plt.ylabel('Frequency');
```

```python
x = rng.normal(100,15,10_000)
y = rng.normal(120,20,10_000)

plt.hist(x,histtype='step',label='Global')
plt.hist(y,histtype='step',label='Asia')
plt.title('Some Graph',fontsize='20',fontweight='bold')
plt.xlabel('IQ Score')
plt.ylabel('Frequency')
plt.legend();
```

**Note:** To make a normalised histogram, set `density` to `True`.

```python
x = rng.uniform(20,250,10_000)
y = rng.normal(120,20,10_000)

plt.hist(x,cumulative=True,histtype='step',label='Global')
plt.hist(y,cumulative=True,histtype='step',label='Asia')
plt.title('Some Cumulative Graph',fontsize='20',fontweight='bold',pad=10)
plt.xlabel('IQ Score')
plt.ylabel('Frequency')
plt.legend();
```

## Pie Chart

```python
x = ['SUVs','Sedans','Hatchbacks','Coupes','Supercars','Hypercars']
y = np.array([50_000,41_000,35_000,20_000,10_000,5_000])

plt.pie(y,labels=x,explode=[0,0,0,0,0.1,0.3],autopct='%.2f');
```

```python
x = ['SUVs','Sedans','Hatchbacks','Coupes','Supercars','Hypercars']
y = np.array([50_000,41_000,35_000,20_000,10_000,5_000])

plt.pie(y,labels=x,wedgeprops=dict(width=0.3));
```

## Stem Plot

```python
x = np.arange(0,50,2)
y = rng.integers(-50,50,len(x))
z = rng.integers(-25,60,len(x))

plt.stem(x,y,linefmt='g--',markerfmt='g*',label='My Business')
plt.stem(x,z,'p-.',label='His Business')
plt.legend();
```

## Step Plot

```python
x = np.arange(0,100,10)
y = np.sin(x)

plt.step(x,y,marker='o',color='purple');
```

## Stack/Area Plot

```python
x = np.arange(1,13)
comp1 = rng.integers(0,100,len(x))
comp2 = rng.integers(50,100,len(x))
comp3 = rng.integers(0,50,len(x))

plt.stackplot(x,comp1,comp2,comp3,labels=('Company 1','Company 2','Company 3'))
plt.legend();
```
